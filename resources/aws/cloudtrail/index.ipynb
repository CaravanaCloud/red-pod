```python
import os
import pandas as pd
import json

datafiles = []
for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith(".json.gz"):
            datafiles.append(os.path.join(root, file))

datafiles
```

```python
# Reading all CSV files into a list of dataframes
dataframes = [pd.read_json(file) for file in datafiles]

# Combining all dataframes into a single dataframe
dff = pd.concat(dataframes, ignore_index=True)
df = [pd.json_normalize(record) for record in dff['Records']]
df = pd.concat(df, ignore_index=True)

df.head()
```

```python
df[df['eventName'] == 'RunInstances'] 
```

```python
pd.set_option('display.max_rows', None)
df['service'] = df['eventSource'].str.replace('.amazonaws.com', '', regex=False)
df['api_call'] = df['service'] + ':' + df['eventName']


api_call_counts = df['api_call'].value_counts().sort_index()

api_call_counts
```

```python
output_file_path_counts = 'api_calls_counts.txt'

# Write the counts of each API call to a file
with open(output_file_path_counts, 'w') as file:
    for api_call, count in api_call_counts.items():
        file.write(f"{api_call}: {count}\n")
```
